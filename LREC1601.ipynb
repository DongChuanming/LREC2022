```python
def common_list(a,b,c):
    lindex=list()
    assert len(a)<=len(b)
    la=list()
    lb=list()
    for m1 in a:
        la.append(m1.lower())
    for m2 in b:
        lb.append(m2.lower())
    for i in range(len(la)):
        m=la[i]
        if m not in c:
            if m in lb:
                lindex.append(i)
                
    if len(lindex) >0:
        return a[min(lindex):max(lindex)+1]
    else:
        return lindex
```

```python
def reverse_token(l, l1,l2):
    sl=' '.join(l)
    assert len(l1)==len(l2)
    for i in range(len(l1)):
        sl=sl.replace(l2[i],l1[i])
    return sl
    
```

```python
def symbol_split(ch,s):
    rec=""
    lr=list()
    for l in ch:
        if l in s:
            rec=rec.strip()
            if len(rec)>0:
                lr.append(rec)
                rec=""
        else:
            rec+=l
    rec=rec.strip()
    if len(rec)>0:
        lr.append(rec)
    return lr
```

```python
def sub_generation(rec,d,k):
    nr=rec.copy()
    for e in rec:
        #ch=nc
        for v in d[k]:
            nc=e.replace(k,v)
            lnc=nc.strip().split(" ")
            while "" in lnc:
                lnc.remove("")
            nr.add(" ".join(lnc))
    return nr
```
